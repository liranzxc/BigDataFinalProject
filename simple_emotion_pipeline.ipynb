```python
import re
from transformers import pipeline

class ToxicityEnsemble:
    def __init__(self, weights=None):
        """
        Initialize the ensemble with three pipelines.

        The weights dictionary maps a short key to the contribution of
        each model’s toxicity signal. Default weights here are an example;
        adjust as needed.
        """
        # Default weights for each pipeline
        if weights is None:
            self.weights = {
                'xlm_emo_t': 0.3,
                'toxic_xlm_roberta': 0.3,
                'multilingual_sentiment': 0.3
            }
        else:
            self.weights = weights

        # Load pipelines; note some pipelines use `return_all_scores=True`
        self.pipeline_emo = pipeline(
            "text-classification",
            model="MilaNLProc/xlm-emo-t",
            return_all_scores=True
        )
        self.pipeline_toxic = pipeline(
            "text-classification",
            model="unitary/multilingual-toxic-xlm-roberta",
            return_all_scores=True
        )
        self.pipeline_sentiment = pipeline(
            "text-classification",
            model="nlptown/bert-base-multilingual-uncased-sentiment",
            return_all_scores=True
        )

    def predict(self, text):
        """
        Predict toxicity using all three models and combine the results
        with a weighted average.

        Args:
            text (str): The input text to analyze.

        Returns:
            dict: A dictionary containing each model's raw outputs and a final
                  weighted toxicity score.
        """
        results = {}

        # Get predictions from each model
        emo_results = self.pipeline_emo(text)
        toxic_results = self.pipeline_toxic(text)
        sentiment_results = self.pipeline_sentiment(text)

        # --- Extract toxicity signals ---
        # 1. From the emotion detection model, we assume that the score for "anger"
        #    can serve as a proxy for toxicity.
        emo_toxic_score = 0.0
        for item in emo_results[0]:
            # Check for variants of anger; adjust the string as needed depending on model
            if item['label'].lower() in ['anger', 'angry']:
                emo_toxic_score = item['score']
                break

        # 2. From the toxicity model (unitary), we extract the score for the label "toxic".
        toxic_score = 0.0
        for item in toxic_results[0]:
            if item['label'].lower() == 'toxic':
                toxic_score = item['score']
                break

        # 3. From the sentiment model, we convert star rating to a toxicity-like score.
        #    Example mapping: lower star (e.g., "1 star") implies high toxicity.
        #    Here we take the label with the highest score.
        best_label = max(sentiment_results[0], key=lambda x: x['score'])
        sentiment_toxic_score = 0.0
        match = re.search(r'(\d)', best_label['label'])
        if match:
            star_rating = int(match.group(1))
            # Map star rating to a score: 1 star -> 1.0 (most toxic), 5 stars -> 0.2 (least toxic)
            sentiment_toxic_score = (6 - star_rating) / 5.0

        # --- Combine the toxicity scores using weighted average ---
        import numpy as np
        ensemble_score = np.max([
            self.weights['xlm_emo_t']  ,
            self.weights['toxic_xlm_roberta']  ,
            self.weights['multilingual_sentiment'] ])


        # Gather all details in the results
        results['xlm_emo_t'] = emo_results
        results['toxic_xlm_roberta'] = toxic_results
        results['multilingual_sentiment'] = sentiment_results
        results['weighted_toxicity_score'] = ensemble_score

        return results
```

```python
ensemble_detector = ToxicityEnsemble()
```

```python
text_example = "boobs"
prediction = ensemble_detector.predict(text_example)
print("Model outputs and weighted toxicity score:")
print(prediction)
```
